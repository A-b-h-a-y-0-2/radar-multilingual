```python
import transformers
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch.nn.functional as F
```

```python
ai_text=[]
with open('llama_ai_fr.txt','r') as f:
    line = f.readline()
    while line:
        line = line.strip()
        line = line.replace('start : ',' ')
        line = line.replace('[\"',' ')
        line = line.replace('[\'',' ')
        line = line.replace('\"]',' ')
        line = line.replace('\']',' ')
        ai_text.append(line)
        print(line)
        line = f.readline()
```

```python

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-fr-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-fr-en")
model.eval()

model.to(torch.device('cuda'))
```

```python
from datasets import load_dataset
```

```python
datawiki_fr = load_dataset('wikipedia', '20220301.fr')
```

```python
human_fr_tr =[]
with open('human_fr_tr.txt', 'r') as f:
    line = f.readline()
    while line :
        line  = line.strip()
        human_fr_tr.append(line)
        line = f.readline()
        
```

```python
for i in range(len(ai_text)):
    ai_text[i] = ai_text[i][:1024]

ai_fr_tr =[]
count = 0
for text in ai_text:
    inp = text[:512]
    input_ids = tokenizer(inp, return_tensors="pt").input_ids
    input_ids = input_ids.to(torch.device('cuda'))
    outputs = model.generate(input_ids=input_ids, num_beams=5, num_return_sequences=1)
    ai_fr_tr.append(tokenizer.batch_decode(outputs, skip_special_tokens=True))
    count+=1
    print(count)
    del input_ids
    torch.cuda.empty_cache()


print('Ai translation done')
```

```python
with open('llama_ai_fr_tr.txt', 'w') as f:
    for line in ai_fr_tr:
        f.write(str(line)+'\n')
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="openai-community/roberta-large-openai-detector")
```

```python
device = "cuda"# example: cuda:0
detector_path_or_id = "TrustSafeAI/RADAR-Vicuna-7B"
detector = transformers.AutoModelForSequenceClassification.from_pretrained(detector_path_or_id)
detector_tokenizer = transformers.AutoTokenizer.from_pretrained(detector_path_or_id)
detector.eval()
detector.to(device)

Text_input = ai_fr_tr
# error = 0
output_probs_list_fr_ai_tr=[]
# Use detector to deternine wehther the text_input is ai-generated.
with torch.no_grad():
  for i  in Text_input:
    inputs = detector_tokenizer(i, padding=True, truncation=True, max_length=512, return_tensors="pt")
    inputs = {k:v.to(device) for k,v in inputs.items()}
    output_probs = F.log_softmax(detector(**inputs).logits,-1)[:,0].exp().tolist()
    output_probs_list_fr_ai_tr.append(output_probs)
# output_probs is the probability that the input_text is generated by LLM.
# for result in output_probs:
#   error +=  result 
# error = error/len(output_probs)
print(len(output_probs_list_fr_ai_tr))
file_ai = "llama_ai_fr_tr.csv"
```

```python
with open(file_ai, 'w') as f:
    for num in output_probs_list_fr_ai_tr:
        f.write(str(num[0]))


```
